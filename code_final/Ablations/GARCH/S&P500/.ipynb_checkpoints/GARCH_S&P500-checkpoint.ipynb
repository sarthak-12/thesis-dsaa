# Relevant Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import seaborn as sns
```

# Preprocessing

```python
filepath = '../../dataset_final/FinSen_S&P500/merged_S&P500.csv'
data = pd.read_csv(filepath)
data.head()
```

```python
if 'Unnamed: 0' in data.columns:
    data = data.drop(columns=['Unnamed: 0'])
data.reset_index
data.head()
```

```python
data.columns
```

```python
# (B) Convert 'Date' to datetime and set as index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date')
data.sort_index(inplace=True)

# (C) Drop rows with any missing values (or handle in another way if desired)
data.dropna(subset=['Daily_Return'], inplace=True)

# (D) (Optional) If you donâ€™t have a column for Daily_Return (log returns), you can create it:
data['Log_Return'] = np.log(data['Close'] / data['Close_lag1'])  # Example if Close_lag1 is the previous close

data.reset_index()

# Inspect the first few rows
data.head()
```

# Stationarity Test

```python
# ==============================================
# 2. QUICK EDA & STATIONARITY TEST ON RETURNS
# ==============================================

# (A) Plot the returns to get a sense of the series
data['Log_Return'].plot(title='Daily Return Over Time', figsize=(10, 4))
plt.show()

# (B) ADF Test (Augmented Dickey-Fuller) for stationarity
adf_result = adfuller(data['Log_Return'].dropna())
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
for key, value in adf_result[4].items():
    print('Critical Values:')
    print(f'   {key}, {value}')
```

```python
# (C) ACF Plot to examine autocorrelations
plt.figure(figsize=(12, 6))
plot_acf(data['Log_Return'].dropna(), lags=30, title="ACF Plot for Log Returns")
plt.show()

# (D) PACF Plot to examine partial autocorrelations
plt.figure(figsize=(12, 6))
plot_pacf(data['Log_Return'].dropna(), lags=30, title="PACF Plot for Log Returns", method='ywm')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Step 1: Fit a mean model (AR(1) for simplicity)
ar_model = ARIMA(data['Log_Return'].dropna(), order=(1, 0, 0)).fit()

# Step 2: Extract residuals
residuals = ar_model.resid

# Step 3: Plot ACF and PACF for residuals
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(residuals, lags=30, title="ACF of Residuals", zero=False)
plt.subplot(122)
plot_pacf(residuals, lags=30, title="PACF of Residuals", zero=False)
plt.tight_layout()
plt.show()

# Step 4: Square the residuals
squared_residuals = residuals ** 2

# Step 5: Plot ACF and PACF for squared residuals
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(squared_residuals, lags=30, title="ACF of Squared Residuals", zero=False)
plt.subplot(122)
plot_pacf(squared_residuals, lags=30, title="PACF of Squared Residuals", zero=False)
plt.tight_layout()
plt.show()
```

### **Observation**
- There are **significant spikes** at multiple lags, particularly for the first few lags (e.g., 1, 2, 3, 4, etc.).
- This indicates that the **squared residuals are autocorrelated**, suggesting **volatility clustering** in the data.

---

### **Interpretation**
- The presence of significant autocorrelations in squared residuals confirms that **variance is time-dependent** and not constant.
- This behavior justifies the use of a **GARCH model** to capture **conditional heteroscedasticity**.

The time series is stationary so, we can proceed.

# Specify and Fit the GARCH model

```python
# Step 2.1: Prepare Data
# Ensure 'Date' is a datetime index and drop NaNs for the relevant columns
data['Date'] = pd.to_datetime(data['Date'])
data.dropna(subset=['Log_Return', 'Volume_lag1', 'aggregate_sentiment_score_lag1'], inplace=True)

# Step 2.2: Define GARCH Model
# Using Daily_Return as the target series, and exogenous variables in the variance equation
returns = data['Log_Return'] * 100  # Scale returns for numerical stability
X_variance = data[['Volume_lag1', 'aggregate_sentiment_score_lag1']]  # Exogenous variables

# Specify GARCH(1,1) with exogenous regressors in variance equation
garch_model = arch_model(
    returns,
    vol='GARCH',
    p=1, q=1,
    mean='Zero',  # Assume zero mean returns; adjust if needed
    x=X_variance,  # Exogenous regressors in variance equation
    dist='normal'  # Error distribution
)

# Step 2.3: Fit the Model
res = garch_model.fit(disp='off')
print(res.summary())
```

# Forecast

```python
# 1. Get the full in-sample forecast
in_sample_forecast = res.forecast(start=0,reindex=False)

# 'h.1' refers to the in-sample variance for horizon 1
cond_var = in_sample_forecast.variance['h.1']
# 6. Then compute volatility and residuals
cond_vol = cond_var ** 0.5
residuals = res.resid

# 7. Finally, attach them to your DataFrame
data['garch_cond_variance'] = cond_var
data['garch_cond_volatility'] = cond_vol
data['garch_residuals'] = residuals.reindex(data.index)
data.head()
```

# Diagnosis and evaluation

## Conditional Volatility Plot

```python
# Ensure 'Date' is in datetime format and set as index (if not already)
data['Date'] = pd.to_datetime(data['Date'])

# Create the Seaborn lineplot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='garch_cond_volatility', data=data)

# Customize the plot
plt.title("Conditional Volatility Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Conditional Volatility", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()
```

## Correlation Metrics

```python
# Select the relevant columns (GARCH features and Movement)
correlation_data = data[['garch_cond_variance', 'garch_cond_volatility', 'garch_residuals', 'Movement']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Display the correlations of Movement with other features
print(correlation_matrix['Movement'])
```

```python
# Heatmap for correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

## Lagged Features

```python
# Define the lags to test
lags = [1, 2, 3, 5, 7]

# Create lagged features for each lag
for lag in lags:
    data[f'garch_cond_variance_lag{lag}'] = data['garch_cond_variance'].shift(lag)
    data[f'garch_cond_volatility_lag{lag}'] = data['garch_cond_volatility'].shift(lag)
    data[f'garch_residuals_lag{lag}'] = data['garch_residuals'].shift(lag)

# Select only lagged features and Movement for correlation analysis
lagged_columns = [f'garch_cond_variance_lag{lag}' for lag in lags] + \
                 [f'garch_cond_volatility_lag{lag}' for lag in lags] + \
                 [f'garch_residuals_lag{lag}' for lag in lags] + ['Movement']

lagged_correlation_data = data[lagged_columns]

# Drop rows with NaNs introduced by shifting
lagged_correlation_data.dropna(inplace=True)

# Calculate the correlation matrix
lagged_correlation_matrix = lagged_correlation_data.corr()

# Extract and display correlations of Movement with all lagged features
print("Correlation of Movement with lagged GARCH features:")
for lag in lags:
    print(f"\nLag {lag} days:")
    print(f"  garch_cond_variance_lag{lag}: {lagged_correlation_matrix.loc['Movement', f'garch_cond_variance_lag{lag}']:.4f}")
    print(f"  garch_cond_volatility_lag{lag}: {lagged_correlation_matrix.loc['Movement', f'garch_cond_volatility_lag{lag}']:.4f}")
    print(f"  garch_residuals_lag{lag}: {lagged_correlation_matrix.loc['Movement', f'garch_residuals_lag{lag}']:.4f}")

# Filter correlations of Movement with lagged features
movement_correlations = lagged_correlation_matrix['Movement'].filter(like='lag')

# Plot the correlations
plt.figure(figsize=(12, 6))
sns.barplot(x=movement_correlations.index, y=movement_correlations.values)
plt.title("Correlation of Movement with Lagged GARCH Features", fontsize=16)
plt.ylabel("Correlation", fontsize=12)
plt.xlabel("Lagged Features", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Rolling Averages

```python
window_sizes = [3, 5, 7]

for window in window_sizes:
    data[f'rolling_cond_variance_{window}'] = data['garch_cond_variance'].rolling(window=window).mean()
    data[f'rolling_cond_volatility_{window}'] = data['garch_cond_volatility'].rolling(window=window).mean()
    data[f'rolling_residuals_{window}'] = data['garch_residuals'].rolling(window=window).mean()

# Display a sample of the rolling features
print(data.filter(like='rolling_').head(10))  # Display all rolling features
```

```python
# Dynamically select all rolling columns and 'Movement'
rolling_columns = [col for col in data.columns if 'rolling_' in col] + ['Movement']

# Calculate and display the correlation matrix
correlation_matrix = data[rolling_columns].corr()

# Print the correlation of 'Movement' with all rolling features
print("Correlation of Movement with rolling GARCH features:")
print(correlation_matrix['Movement'].sort_values(ascending=False))
```

## Causality Testing

```python
from statsmodels.tsa.stattools import grangercausalitytests

# Example: Test Granger causality for 1 to 5 lags
features_to_test = ['rolling_residuals_3', 'rolling_cond_variance_3', 'rolling_cond_volatility_3', 'rolling_residuals_5', 'rolling_cond_variance_5', 'rolling_cond_volatility_5', 'rolling_residuals_7', 'rolling_cond_variance_7', 'rolling_cond_volatility_7']
max_lags = 5

for feature in features_to_test:
    print(f"\nTesting Granger Causality: {feature} -> Movement")
    causality_data = data[['Movement', feature]].dropna()  # Drop NaNs
    grangercausalitytests(causality_data, maxlag=max_lags, verbose=True)
```

### **Summary of Granger Causality Results**

- **`rolling_cond_volatility_3` and `rolling_cond_volatility_5` â†’ Movement**:  
  Significant causality detected at **lag 1** (p-values: 0.0486, 0.0489), suggesting short-term smoothed volatility influences `Movement`.

- **`rolling_residuals_3`, `rolling_residuals_5`, `rolling_residuals_7` â†’ Movement**:  
  No significant causality found across all lags, indicating limited predictive power.

- **`rolling_cond_variance_3`, `rolling_cond_variance_5`, `rolling_cond_variance_7` â†’ Movement**:  
  Weak evidence of causality (lowest p-values ~0.09), but none below 0.05, suggesting minimal contribution.

- **Lag Dependency**:  
  Significant effects observed primarily at **lag 1**, highlighting short-term relationships.

- **Feature Prioritization**:  
  Focus on **`rolling_cond_volatility_3` and `rolling_cond_volatility_5` (lag 1)** for further modeling due to their strongest and most consistent causal relationships.
